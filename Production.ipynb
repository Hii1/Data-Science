```python
#19290116
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset\Production.csv")[["Area", "Year", "Value"]]
df.rename({"Value": "Production (tonnes)", "Area": "Country"}, axis = 1, inplace = True)
df = df[(df["Country"] == "Turkey") | 
   (df["Country"] == "Palestine") |
   (df["Country"] == "United States of America") |
   (df["Country"] == "Canada") |
   (df["Country"] == "Japan")
  ]## you can remove this statmetn if you want all acountries to be included.
df.reset_index(inplace=True)
del df["index"]
```

```python
#Production 2000-2019 by country
df.to_csv("result\Production_5Countries.csv")
df
```

```python
plt.figure(figsize=(20,10)) 
for group, farm in df.groupby("Country"):
    plt.plot(farm["Year"], farm["Production (tonnes)"], label = group)

plt.legend(loc = (1,0))
plt.title("Production of cabbage 2000-2019")
plt.xlabel("Year")
plt.ylabel("Production (Tonnes)")
plt.xticks(np.arange(2000,2020))
plt.savefig("result\Production1.png")
plt.show()
```

```python
new_df = df[df["Year"] == 2000]
plt.figure(figsize=(20,10)) 

plt.bar(new_df["Country"], new_df["Production (tonnes)"])
plt.title("Production of cabbage 2000")
plt.xlabel("Country")
plt.ylabel("Production (Tonnes)")
plt.savefig("result\Production2.png")
plt.show()
```
