```python
#19290116
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset\Area_harvested.csv")[["Area", "Year", "Value"]]
df.rename({"Value": "Area ha", "Area": "Country"}, axis = 1, inplace = True)
df = df[(df["Country"] == "Turkey") | 
   (df["Country"] == "Palestine") |
   (df["Country"] == "United States of America") |
   (df["Country"] == "Canada") |
   (df["Country"] == "Japan")
  ] ## you can remove this statmetn if you want all acountries to be included.
df.reset_index(inplace=True)
del df["index"]
```

```python
#Area_harvested 2000-2019 by country
df.to_csv("result\Area_harvested_5Countries.csv")
df
```

```python
plt.figure(figsize=(20,10)) 
for group, farm in df.groupby("Country"):
    plt.plot(farm["Year"], farm["Area ha"], label = group)

plt.legend(loc = (1,0))
plt.title("Area harvested of cabbage 2000-2019")
plt.xlabel("Year")
plt.ylabel("Area harvested (ha)")
plt.xticks(np.arange(2000,2020)) 
#plt.savefig("result\Area_harvested1.png")
plt.show()
```

```python
new_df = df[df["Year"] == 2000]
plt.figure(figsize=(20,10)) 

plt.bar(new_df["Country"], new_df["Area ha"])
plt.title("Area harvested of cabbage 2000")
plt.xlabel("Country")
plt.ylabel("Area harvested (ha)")
#plt.savefig("result\Area_harvested2.png")
plt.show()
```
