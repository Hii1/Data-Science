```python
#19290116
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset\Yield.csv")[["Area", "Year", "Value"]]
df.rename({"Value": "Yield (hg/ha)", "Area": "Country"}, axis = 1, inplace = True)
df = df[(df["Country"] == "Turkey") | 
   (df["Country"] == "Palestine") |
   (df["Country"] == "United States of America") |
   (df["Country"] == "Canada") |
   (df["Country"] == "Japan")
  ]## you can remove this statmetn if you want all acountries to be included.
df.reset_index(inplace=True)
del df["index"]
```

```python
#Yield 2000-2019 by Country
df.to_csv("result\Yield_5Countries.csv")
df
```

```python
plt.figure(figsize=(20,10)) 
for group, farm in df.groupby("Country"):
    plt.plot(farm["Year"], farm["Yield (hg/ha)"], label = group)

plt.legend(loc = (1,0))
plt.title("Yield of cabbage 2000-2019")
plt.xlabel("Year")
plt.ylabel("Yield (hg/ha)")
plt.xticks(np.arange(2000,2020)) 
plt.savefig("result\Yield1.png")
plt.show()
```

```python
new_df = df[df["Year"] == 2000]
plt.figure(figsize=(20,10)) 

plt.bar(new_df["Country"], new_df["Yield (hg/ha)"])
plt.title("Area harvested of cabbage 2000")
plt.xlabel("Country")
plt.ylabel("Yield (hg/ha)")
plt.savefig("result\Yield2.png")
plt.show()
```
